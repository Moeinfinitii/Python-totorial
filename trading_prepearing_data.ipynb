# Import libraries

```python
import pandas as pd 
```

```python
data = pd.read_csv('trading.csv')
```

```python
data.head()
```

```python
data['Tech Rating'][0] in ['Sell','Strong sell',]
```

```python
sell =0
buy = 0
non = 0

for i in range (len(data)):
    if data['Tech Rating'][i] in ['Sell','Strong sell',]:
        data['Tech Rating'][i] = +1
        sell += 1
    elif data['Tech Rating'][i] in ['Buy', 'Strong Buy']:
        data['Tech Rating'][i] = -1
        buy += 1
    else:
        data = data.drop(index=i)
        non +=1
data = data.drop(['Symbol'],axis = 1)
data = data.reset_index(drop=True)
data
```

```python
print('Number of sell = ', sell)
print('Number of buy = ', buy)
print('Number of non = ', non)
```

```python
df = pd.DataFrame(data)

# Export DataFrame to CSV
df.to_csv('trading_prepeard.csv', index=False)
```
