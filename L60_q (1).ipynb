```python
class Card:
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
    def __str__(self):
      cards=f'{self.rank} of {self.suit}'
      return str(cards)

card=Card('Hearts','Two')
print(card)
```

ساختن کلاس دسته

```python
import random
class Deck:
    def __init__(self):
      self.ranks=ranks
      self.suits=suits
      self.cardslist=[]

    def make_cards(self):
       for suit in suits:
          for rank in ranks:
            card=Card(suit,rank)
            self.cardslist.append(str(card))
       return '\n'.join(self.cardslist)



    def shuffle(self):
      return random.shuffle(self.cardslist)

    def deal(self):
      return self.cardslist.pop()

mylist=Deck()
#print(mylist)
mylist.make_cards()
mylist.shuffle()
mylist.deal()





```
